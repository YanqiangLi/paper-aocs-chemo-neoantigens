```python
import pandas, seaborn, scipy, numpy

%matplotlib inline
from matplotlib import pyplot
```

```python
sources = pandas.read_csv("../data/derived/sources.csv", index_col="source_id")
sources = sources.ix[sources.cohort == "AOCS"]
#sources
```

```python
(~sources.treated).sum(), (sources.treated).sum(), sources.donor.nunique()
```

```python
mutations = pandas.read_csv("../data/mutations.csv.bz2")
```

```python
mutations["interesting"] = (mutations.context_mutation_3p5p == "C(C>A)C").astype(float)
mutations["interesting"].mean()
```

```python
mutations.ix[mutations.interesting]
```

```python
samples_with_rna = mutations.source_id.ix[mutations.rna_total_reads > 0].unique()
samples_with_rna
```

```python
sources["protein changing"] = mutations.ix[
    (
        (mutations.called == 1) |
        (
            (mutations.dna_alt_reads > 10)
        )
    )
    & (mutations.effect.str.startswith("p."))
].groupby("source_id").donor.count().to_frame()["donor"]

sources["neoantigens"] = mutations.ix[
    (
        (mutations.called == 1) |
        (
            (mutations.dna_alt_reads > 10)
        )
    )
    & (mutations.binding_affinity < 500)
].groupby("source_id").donor.count().to_frame()["donor"]

sources["expressed neoantigens"] = mutations.ix[
    (
        (mutations.called == 1) |
        (
            (mutations.dna_alt_reads > 10)
        )
    )
    & (mutations.binding_affinity < 500) & (mutations.rna_alt_reads > 2)
].groupby("source_id").donor.count().to_frame()["donor"]

sources["mutations"] = mutations.ix[
    (
        (mutations.called == 1) |
        (
            (mutations.dna_alt_reads > 10)
        )
    )
].groupby("source_id").donor.count().to_frame()["donor"]

sources["interesting mutations"] = mutations.ix[
    (
        (mutations.called == 1) |
        (
            (mutations.dna_alt_reads > 10)
        )
    )
].groupby("source_id").interesting.mean().to_frame()["interesting"] * 100.0



sources
```

```python
def mean_with_std(grouped_column, decimals=0):
    pattern = "%%0.%df" % decimals
    return [
        (pattern + " +/ " + pattern) % (m, s) if not pandas.isnull(s) else pattern % m
        for (m, s) in zip(grouped_column.mean(), grouped_column.std())
    ]


grouping_columns = ["cohort", "library", "tissue_type", "treated"]
grouped = sources.groupby(grouping_columns)

df = grouped.donor.nunique().to_frame() #.reset_index()
df["samples"] = grouped.donor.count().to_frame()["donor"]
df["samples with RNA"] = sources.ix[sources.index.isin(samples_with_rna)].groupby(grouping_columns).donor.count().to_frame()["donor"]
df["Mutations"] = mean_with_std(grouped.mutations)
df["C(C>A)C percent"] = mean_with_std(grouped["interesting mutations"], decimals=2)
df["Protein changing"] = mean_with_std(grouped["protein changing"])
df["Neoantigenic"] = mean_with_std(grouped.neoantigens)
df["Expressed neoantigenic"] = mean_with_std(grouped["expressed neoantigens"])


df = df.reset_index()
df["treated"] = df.treated.map({False: "untreated", True: "treated"})
df["samples"] = ["%d (%d)" % (row.samples, row["samples with RNA"]) for (i, row) in df.iterrows()]
del df["samples with RNA"]
df = df.rename(columns={
        "donor": "donors",
        "treated": "treatment",
        "samples": "samples (with RNA)",
        "tissue_type": "tissue type",
    })

df.to_csv("../data/cohort.csv")
df
```
